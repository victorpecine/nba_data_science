```python
df = df_stats[(df_stats['PLAYER_NAME'] == 'LeBron James') | (df_stats['PLAYER_NAME'] == 'Aaron Gordon') | (df_stats['PLAYER_NAME'] == 'Zaza Pachulia')].reset_index(drop=True)
df.head()
```

```python
df_reversed = df.iloc[::-1].reset_index(drop=True)
df_reversed.head()
```

```python
fig = px.line(df_reversed, x='SEASON', y='FG_PCT', color='NICKNAME')
fig.update_xaxes(type='category')

fig.show()
```
